# **Monte Carlo Simulation for Project Risk Management**


---



Description:

This Python script implements a Monte Carlo simulation for project risk management. It helps in estimating project contingencies by modeling uncertain parameters (such as cost, time, and resources) and running multiple simulations to predict possible outcomes. The tool also generates graphs to visualize the probability distribution of the results, helping decision-makers better understand potential risks and plan accordingly.

Key features:

    Monte Carlo Simulation: Simulates different scenarios based on uncertain project parameters.
    Contingency Calculation: Estimates the required contingency based on the distribution of possible outcomes.
    Visualization: Generates histograms and cumulative probability plots for clear result interpretation.

Output:

    A list of possible project costs or durations.
    Estimated contingency percentage based on risk tolerance.
    Visual representations of the simulation results (e.g., histograms, cumulative probability charts).

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate the project based on uncertain parameters
def monte_carlo_simulation(n_simulations, cost_mean, cost_std, time_mean, time_std):
    project_costs = []
    project_durations = []

    for _ in range(n_simulations):
        # Generate random values for cost and duration based on normal distribution
        cost = np.random.normal(cost_mean, cost_std)
        duration = np.random.normal(time_mean, time_std)

        # Append results to lists
        project_costs.append(cost)
        project_durations.append(duration)

    return project_costs, project_durations

# Function to calculate contingency based on simulations
def calculate_contingency(results, risk_tolerance=0.95):
    # Calculate the contingency based on risk tolerance
    contingency = np.percentile(results, (1 - risk_tolerance) * 100)
    return contingency

# Function to plot results
def plot_results(project_costs, project_durations):
    # Plot the histogram of costs
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(project_costs, bins=50, color='blue', alpha=0.7)
    plt.title('Project Cost Distribution')
    plt.xlabel('Cost')
    plt.ylabel('Frequency')

    # Plot the histogram of durations
    plt.subplot(1, 2, 2)
    plt.hist(project_durations, bins=50, color='green', alpha=0.7)
    plt.title('Project Duration Distribution')
    plt.xlabel('Duration')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Main execution function
def main():
    # Define the simulation parameters
    n_simulations = 10000
    cost_mean = 100000  # Example: project mean cost
    cost_std = 15000    # Example: project cost standard deviation
    time_mean = 24      # Example: project mean duration in months
    time_std = 4        # Example: project duration standard deviation

    # Run the Monte Carlo simulation
    project_costs, project_durations = monte_carlo_simulation(
        n_simulations, cost_mean, cost_std, time_mean, time_std
    )

    # Calculate contingency for cost and time
    cost_contingency = calculate_contingency(project_costs)
    time_contingency = calculate_contingency(project_durations)

    # Output the results
    print(f"Cost Contingency (95% confidence): ${cost_contingency:.2f}")
    print(f"Time Contingency (95% confidence): {time_contingency:.2f} months")

    # Plot the results
    plot_results(project_costs, project_durations)

# Run the script
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate a single activity based on PERT distribution
def pert_simulation(optimistic, most_probable, pessimistic, n_simulations):
    # Calculate the expected cost (mean) and standard deviation based on PERT
    expected_cost = (optimistic + 4 * most_probable + pessimistic) / 6
    std_dev = (pessimistic - optimistic) / 6

    # Simulate the cost using a normal distribution based on expected cost and standard deviation
    simulated_costs = np.random.normal(expected_cost, std_dev, n_simulations)
    return simulated_costs

# Function to calculate contingency based on simulations
def calculate_contingency(results, confidence_level=0.9):
    # Calculate the contingency based on the given confidence level
    contingency = np.percentile(results, (1 - confidence_level) * 100)
    return contingency

# Function to plot the results
def plot_results(all_costs):
    plt.hist(all_costs, bins=50, color='blue', alpha=0.7)
    plt.title('Project Cost Distribution')
    plt.xlabel('Cost')
    plt.ylabel('Frequency')
    plt.show()

# Main execution function
def main():
    # Define the simulation parameters
    n_simulations = 10000

    # List of activities with (Optimistic, Most Probable, Pessimistic) costs
    activities = [
        {"name": "Activity 1", "optimistic": 50000, "most_probable": 70000, "pessimistic": 90000},
        {"name": "Activity 2", "optimistic": 30000, "most_probable": 50000, "pessimistic": 80000},
        {"name": "Activity 3", "optimistic": 20000, "most_probable": 35000, "pessimistic": 60000},
        {"name": "Activity 4", "optimistic": 40000, "most_probable": 60000, "pessimistic": 85000},
        {"name": "Activity 5", "optimistic": 35000, "most_probable": 50000, "pessimistic": 75000},
        {"name": "Activity 6", "optimistic": 45000, "most_probable": 65000, "pessimistic": 90000},
        {"name": "Activity 7", "optimistic": 25000, "most_probable": 40000, "pessimistic": 60000},
        {"name": "Activity 8", "optimistic": 15000, "most_probable": 30000, "pessimistic": 45000},
        {"name": "Activity 9", "optimistic": 60000, "most_probable": 80000, "pessimistic": 100000},
        {"name": "Activity 10", "optimistic": 55000, "most_probable": 75000, "pessimistic": 95000},
    ]

    # List to store the results from all activities
    total_project_costs = []

    # Run Monte Carlo simulation for each activity and accumulate the total costs
    for activity in activities:
        costs = pert_simulation(
            activity["optimistic"],
            activity["most_probable"],
            activity["pessimistic"],
            n_simulations
        )
        total_project_costs.extend(costs)  # Accumulate the costs of all activities

    # Calculate the contingency for the total project cost
    project_contingency = calculate_contingency(total_project_costs, confidence_level=0.9)

    # Output the results
    print(f"Cost Contingency (90% confidence): ${project_contingency:.2f}")

    # Plot the total project cost distribution
    plot_results(total_project_costs)

# Run the script
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# =========================================================
# Funcin para simular una sola actividad con distribucin PERT
# =========================================================
def pert_simulation(optimistic, most_probable, pessimistic, n_simulations):
    """
    Genera simulaciones de costo para una actividad usando una aproximacin
    de la distribucin PERT basada en una distribucin normal, tomando:
      - La media como (optimistic + 4*most_probable + pessimistic)/6
      - La desviacin estndar como (pessimistic - optimistic)/6

    Retorna un array de numpy con n_simulations valores simulados.
    """
    # Clculo de la media y la desviacin estndar (PERT aproximado con normal)
    expected_cost = (optimistic + 4 * most_probable + pessimistic) / 6
    std_dev = (pessimistic - optimistic) / 6

    # Generar muestras de una distribucin normal
    simulated_costs = np.random.normal(expected_cost, std_dev, n_simulations)
    return simulated_costs

# =========================================================
# Funcin para calcular un percentil dado de los resultados
# =========================================================
def calculate_percentile(results, percentile=90):
    """
    Retorna el valor en 'results' correspondiente al percentile dado.
    Por ejemplo, percentile=90 -> P90.
    """
    return np.percentile(results, percentile)

# =========================================================
# Funcin para graficar el histograma de costos totales
# con un sombreado para el intervalo de confianza
# =========================================================
def plot_total_distribution(total_costs, confidence=0.90, simple_sum=None):
    """
    Genera un histograma de la distribucin de costos totales y muestra:
      - El intervalo de confianza (IC) al 'confidence' (ej. 90%) sombreado.
      - Una lnea vertical que indique la "suma simple" (opcional).
    """
    plt.figure(figsize=(8, 6))

    # Histograma principal
    counts, bins, patches = plt.hist(total_costs, bins=50, color='blue', alpha=0.7, edgecolor='black')

    # Clculo de lmites para el IC al 90% (ej. 5 y 95 percentil para un 90% central)
    lower_bound = np.percentile(total_costs, (1 - confidence)/2 * 100)  # 5%
    upper_bound = np.percentile(total_costs, (1 + confidence)/2 * 100)  # 95%

    # Sombrear el intervalo [lower_bound, upper_bound]
    plt.axvspan(lower_bound, upper_bound, color='orange', alpha=0.2, label=f'IC {int(confidence*100)}%')

    # Lnea vertical para la suma simple de las medias (opcional)
    if simple_sum is not None:
        plt.axvline(simple_sum, color='red', linestyle='--', linewidth=2,
                    label=f'Suma simple de medias: {simple_sum:,.0f}')

    # Aadir etiquetas y leyenda
    plt.title('Distribucin del Costo Total del Proyecto')
    plt.xlabel('Costo')
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.tight_layout()
    plt.show()

# =========================================================
# Funcin para graficar la distribucin de cada actividad en subplots
# =========================================================
def plot_activity_distributions(all_activities_costs, activities):
    """
    Muestra un conjunto de subplots (2 x 5 si hay 10 actividades)
    con la distribucin individual de cada actividad.
    """
    n_activities = len(activities)
    cols = 5
    rows = (n_activities + cols - 1) // cols  # Calcular filas necesarias

    plt.figure(figsize=(cols*4, rows*3))

    for i in range(n_activities):
        plt.subplot(rows, cols, i+1)
        plt.hist(all_activities_costs[i], bins=30, color='green', alpha=0.7, edgecolor='black')

        # Calculamos la media PERT para mostrarla con una lnea vertical
        optimistic = activities[i]["optimistic"]
        most_probable = activities[i]["most_probable"]
        pessimistic = activities[i]["pessimistic"]
        mean_pert = (optimistic + 4*most_probable + pessimistic) / 6

        plt.axvline(mean_pert, color='red', linestyle='--', linewidth=2,
                    label=f'Media PERT={mean_pert:.0f}')

        plt.title(f'{activities[i]["name"]}')
        plt.xlabel('Costo')
        plt.ylabel('Frecuencia')
        plt.legend(fontsize=8)

    plt.tight_layout()
    plt.show()

# =========================================================
# Funcin principal
# =========================================================
def main():
    # Semilla para reproducibilidad (opcional)
    np.random.seed(12345)

    # Definir el nmero de simulaciones
    n_simulations = 10000

    # Lista de actividades con (Optimistic, Most Probable, Pessimistic)
    activities = [
        {"name": "Activity 1",  "optimistic": 50000, "most_probable": 70000, "pessimistic": 90000},
        {"name": "Activity 2",  "optimistic": 30000, "most_probable": 50000, "pessimistic": 80000},
        {"name": "Activity 3",  "optimistic": 20000, "most_probable": 35000, "pessimistic": 60000},
        {"name": "Activity 4",  "optimistic": 40000, "most_probable": 60000, "pessimistic": 85000},
        {"name": "Activity 5",  "optimistic": 35000, "most_probable": 50000, "pessimistic": 75000},
        {"name": "Activity 6",  "optimistic": 45000, "most_probable": 65000, "pessimistic": 90000},
        {"name": "Activity 7",  "optimistic": 25000, "most_probable": 40000, "pessimistic": 60000},
        {"name": "Activity 8",  "optimistic": 15000, "most_probable": 30000, "pessimistic": 45000},
        {"name": "Activity 9",  "optimistic": 60000, "most_probable": 80000, "pessimistic": 100000},
        {"name": "Activity 10", "optimistic": 55000, "most_probable": 75000, "pessimistic": 95000},
    ]

    # =========================================================
    # 1) Simular costos para cada actividad y guardar resultados
    # =========================================================
    all_activities_costs = []  # Aqu guardamos array de costos para cada actividad
    for activity in activities:
        costs = pert_simulation(
            activity["optimistic"],
            activity["most_probable"],
            activity["pessimistic"],
            n_simulations
        )
        all_activities_costs.append(costs)

    # Convertir la lista en un array de shape (n_activities, n_simulations)
    all_activities_costs = np.array(all_activities_costs)

    # =========================================================
    # 2) Graficar la distribucin de cada actividad (subplots)
    # =========================================================
    plot_activity_distributions(all_activities_costs, activities)

    # =========================================================
    # 3) Obtener la distribucin de costos totales
    #    sumando los costos de todas las actividades en cada corrida
    # =========================================================
    # Para cada simulacin, sumamos los costos de todas las actividades:
    total_project_costs = np.sum(all_activities_costs, axis=0)  # shape (n_simulations,)

    # =========================================================
    # 4) Calcular la "suma simple" de las medias PERT
    #    (es decir, sin simulacin, slo la suma de los valores esperados)
    # =========================================================
    simple_sum = 0.0
    for activity in activities:
        mean_pert = (activity["optimistic"]
                     + 4 * activity["most_probable"]
                     + activity["pessimistic"]) / 6
        simple_sum += mean_pert

    # =========================================================
    # 5) Calcular percentiles y contingencia
    # =========================================================
    p50 = calculate_percentile(total_project_costs, 50)
    p90 = calculate_percentile(total_project_costs, 90)
    p10 = calculate_percentile(total_project_costs, 10)

    # Por convencin, a veces la "contingencia al 90%" se interpreta como el valor de P90
    # (tal que hay 90% de probabilidad de no excederlo). Ajusta segn tu criterio:
    contingency_90 = p90

    # =========================================================
    # 6) Mostrar resultados numricos
    # =========================================================
    print(f"\n================= RESULTADOS =================")
    print(f"Suma simple (slo medias PERT)         : ${simple_sum:,.2f}")
    print(f"Mediana (P50) de la simulacin total   : ${p50:,.2f}")
    print(f"P10  de la simulacin total            : ${p10:,.2f}")
    print(f"P90  de la simulacin total            : ${p90:,.2f}")
    print(f"Contingencia (usando P90 como referencia): ${contingency_90:,.2f}")
    print("=============================================\n")

    # =========================================================
    # 7) Graficar la distribucin total con el IC al 90%
    #    y la lnea de la suma simple
    # =========================================================
    plot_total_distribution(
        total_costs=total_project_costs,
        confidence=0.90,
        simple_sum=simple_sum
    )

# =========================================================
# Ejecutar el script
# =========================================================
if __name__ == "__main__":
    main()
```
